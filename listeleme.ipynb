```python
import os
import sqlite3

def sil_tablosu_olmayan_veritabanlari(database_folder):
    # Klasördeki tüm dosyaları al
    database_files = [f for f in os.listdir(database_folder) if f.endswith(".db")]

    # Veritabanlarını kontrol edip tablosu olmayanları silen kod
    for db_file in database_files:
        db_path = os.path.join(database_folder, db_file)
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()

        # Tablo sayısını kontrol et
        cursor.execute("SELECT count(*) FROM sqlite_master WHERE type='table';")
        table_count = cursor.fetchone()[0]

        if table_count == 0:
            # Tablo sayısı sıfırsa veritabanını sil
            conn.close()
            os.remove(db_path)
            print(f"{db_file} veritabanı silindi, çünkü tablo bulunmuyordu.")
        else:
            conn.close()
            print(f"{db_file} veritabanında tablo bulundu, silinmedi.")

# Fonksiyonu çağır
sil_tablosu_olmayan_veritabanlari("databases")
```

```python
import os
import sqlite3

# Veritabanı klasörü
database_folder = "databases"
regmodel_db_path = os.path.join("models", "regmodel.db")

# Regmodel veritabanına bağlantı
if os.path.exists(regmodel_db_path):
    regmodel_conn = sqlite3.connect(regmodel_db_path)
    regmodel_cursor = regmodel_conn.cursor()
else:
    regmodel_conn = sqlite3.connect(regmodel_db_path)
    regmodel_cursor = regmodel_conn.cursor()

    # models tablosunu oluştur
    regmodel_cursor.execute('''
        CREATE TABLE IF NOT EXISTS models (
            database_name TEXT,
            table_name TEXT,
            n REAL,
            combo TEXT
        );
    ''')

# Klasördeki tüm dosyaları al
database_files = [f for f in os.listdir(database_folder) if f.endswith(".db")]

# Veritabanlarını tarayıp models tablosuna kaydet
for db_file in database_files:
    db_path = os.path.join(database_folder, db_file)
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Tablo isimlerini al
    tables = cursor.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
    table_names = [table[0] for table in tables]

    for table_name in table_names:
        # Veritabanı ve tablo var mı kontrol et
        regmodel_cursor.execute('''
            SELECT COUNT(*) FROM models
            WHERE database_name = ? AND table_name = ?;
        ''', (db_file, table_name))
        existing_entry_count = regmodel_cursor.fetchone()[0]

        # Eğer yoksa, veritabanı adı, tablo adı, 'n' ve combo değerlerini kaydet
        if existing_entry_count == 0:
            # Tablonun gözlem sayısını al
            cursor.execute(f"SELECT count(*) FROM {table_name};")
            row_count = cursor.fetchone()[0]

            values = (db_file, table_name, row_count, "0")
            regmodel_cursor.execute('''
                INSERT INTO models (database_name, table_name, n, combo)
                VALUES (?, ?, ?, ?);
            ''', values)

    # Veritabanı bağlantısını kapat
    conn.close()

# Commit ve bağlantıyı kapat
regmodel_conn.commit()
regmodel_conn.close()

print("Veri başarıyla kaydedildi.")
```

```python
import os
import sqlite3

# databases klasöründeki tüm dosyaları al
database_folder = "databases"
database_files = [f for f in os.listdir(database_folder) if f.endswith(".db")]

# Veritabanlarındaki tablo isimlerini ve veri tiplerini yazdır
for database_file in database_files:
    db_path = os.path.join(database_folder, database_file)
    print(f"Veritabanı: {database_file}")

    # SQLite veritabanını aç
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Tabloların isimlerini al
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    # Her bir tablonun şemasını al
    for table in tables:
        table_name = table[0]
        print(f"\nTablo: {table_name}")

        # Tablonun şemasını al
        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = cursor.fetchall()

        # Her sütunun ismini ve veri tipini yazdır
        for column in columns:
            column_name = column[1]
            data_type = column[2]
            print(f"{column_name}: {data_type}")

    # Bağlantıyı kapat
    conn.close()
    print("\n" + "=" * 40 + "\n")
```
